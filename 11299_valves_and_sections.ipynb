```python
import numpy as np

from matplotlib import pyplot as plt 
from matplotlib import figure 
```

```python
def v_air(G, S):
  # G - consumption, m3/h 
  # S - area, m2
  return round(G / 3600 / S, 2)


def pipeG(Q, t1=130, t2=70):
  """
  return consuption for pipe in t/h
  Q - Gcal/h - power,
  t1 - high temperature, 
  t2 - low temperature
  """
  G = round(1000* Q / (t1 - t2), 2)
  return G
```

# it module valves

```python
#it module 
  
G_air_intake = 800000 #m3/h  приточный воздух
G_air_recircle = G_air_intake * 0.75 #m3/h  рециркуляционный воздух
G_air_exhaust = G_air_intake

# Клапан приток 
valve_intake_height, valve_intake_width = 2.400, 1.45

# Клапан рециркуляция 
valve_recircle_height, valve_recircle_width = 2.80, 0.8

# Клапан вытяжной
valve_exhaust_height, valve_exhaust_width = 2.600, 1.1


# сечения
live_sec = 1

# Клапан приток 
valve_intake_S = valve_intake_height * valve_intake_width
valve_intake_S_clear = valve_intake_S * live_sec #живое сечение
valve_intake_count = 8 #8 ВУ
valve_intake_count_ahu = 2 #в каждой ВУ 2 клапана 
valve_intake_velocity = v_air(G_air_intake / valve_intake_count, valve_intake_count_ahu*valve_intake_S_clear)

print("Клапаны приточные")
print(f"Количество приточных клапанов всего - {valve_intake_count * valve_intake_count_ahu} шт.")
print(f"Количество приточных клапанов одной ВУ - {valve_intake_count_ahu} шт.")
print(f"Ширина приточного клапана - {valve_intake_width} м., высота - {valve_intake_height} м.")
print(f"S одного клапана приточного - {valve_intake_S} m2.")
print(f"S одного клапана приточного 'живое' (0,7) - {valve_intake_S_clear} m2.")
print(f"S общее клапанов приточных одной ВУ - {valve_intake_count_ahu * valve_intake_S_clear} m2.")
print(f"Расход приточного воздуха общий - {G_air_intake} m3/h")
print(f"Расход приточного воздуха одной ВУ - {G_air_intake/valve_intake_count} m3/h")
print(f"Скорость воздуха клапаны приточные - {valve_intake_velocity} m/s \n")


# Клапан рециркуляция 
valve_recircle_S = valve_recircle_height * valve_recircle_width
valve_recircle_S_clear = valve_recircle_S * live_sec #живое сечение
valve_recircle_count = 8 #8 ВУ
valve_recircle_count_ahu = 2 #в каждой ВУ 2 клапана 
valve_recircle_velocity = v_air(G_air_recircle / valve_recircle_count, valve_recircle_S_clear*valve_recircle_count_ahu)


print("Клапаны рециркуляции")
print(f"Количество клапанов рециркуляции всего - {valve_recircle_count * valve_recircle_count_ahu} шт.")
print(f"Количество клапанов рециркуляции одной ВУ - {valve_recircle_count_ahu} шт.")
print(f"Ширина клапана рециркуляции - {valve_recircle_width} м., высота - {valve_recircle_height} м.")
print(f"S одного клапана рециркуляции - {valve_recircle_S} m2.")
print(f"S одного клапана рециркуляции 'живое' (0,7) - {valve_recircle_S_clear} m2.")
print(f"S общее клапанов рециркуляции одной ВУ - {valve_recircle_S_clear * valve_recircle_count_ahu} m2.")
print(f"Расход рециркуляции воздуха общий - {G_air_recircle} m3/h")
print(f"Расход рециркуляции воздуха одной ВУ - {G_air_recircle/valve_recircle_count} m3/h")
print(f"Скорость воздуха клапан рециркуляции - {valve_recircle_velocity} m/s \n")


# Клапан вытяжной
valve_exhaust_S = valve_exhaust_height * valve_exhaust_width
valve_exhaust_S_clear = valve_exhaust_S * live_sec  #живое сечение
valve_exhaust_count = 8
valve_exhaust_count_ahu = 2 #в каждой ВУ 2 клапана 
valve_exhaust_velocity = v_air(G_air_exhaust / valve_exhaust_count, valve_exhaust_S_clear * valve_exhaust_count_ahu)


print("Клапаны вытяжные")
print(f"Количество клапанов вытяжной всего - {valve_exhaust_count * valve_exhaust_count_ahu} шт.")
print(f"Количество клапанов рециркуляции одной ВУ - {valve_exhaust_count_ahu} шт.")
print(f"Ширина клапана вытяжной - {valve_exhaust_width} м., высота - {valve_exhaust_height} м.")
print(f"S одного клапана вытяжной - {valve_exhaust_S} m2.")
print(f"S одного клапана вытяжной 'живое' (0,7) - {valve_exhaust_S_clear} m2.")
print(f"S общее клапанов вытяжной одной ВУ - {valve_exhaust_S_clear * valve_exhaust_count_ahu} m2.")
print(f"Расход вытяжной воздуха всего - {G_air_exhaust} m3/h")
print(f"Расход вытяжной воздуха одной ВУ - {G_air_exhaust / valve_exhaust_count} m3/h")
print(f"Скорость воздуха клапан вытяжной - {valve_exhaust_velocity} m/s \n")


print("----------------------------------------------------------------------------------")
```

# Sections

```python

def v_air(G, S):
  # G - consumption, m3/h 
  # S - area, m2
  return round(G / 3600 / S, 2)


# air_intake
G = G_sec_0 = 800000

G_sec_1 = 7 * (G/8)
G_sec_2 = 5 * (G/8)
G_sec_3 = 4 * (G/8)
G_sec_4 = 3 * (G/8)
G_sec_5 = G/8
G_sec_6 = G/8

# air_exhaust
G_sec_8 = 2 * (G/8)
G_sec_9 = 3 * (G/8)
G_sec_10 = 4 * (G/8)
G_sec_11 = 5 * (G/8)
G_sec_12 = 6 * (G/8)
G_sec_13 = 7 * (G/8)
G_sec_14 = G


print(f"Приточный воздух - расход общий - {G} m3/h \n ")

#section areas 
intake_coridor_width = 2.
intake_coridor_height = 2.45

section_0 = round(9.4 * 2.45 + 17.9 * 4.0, 2)
section_1 = round(intake_coridor_width * intake_coridor_height + 13.6 * 4.5, 2)
section_2 = round(intake_coridor_width * intake_coridor_height + 11.5 * 4.5, 2)
section_3 = round(intake_coridor_width * intake_coridor_height + 11.5 * 4.7, 2)
section_4 = round(intake_coridor_width * intake_coridor_height + 11.5 * 4.7, 2)
section_5 = round(intake_coridor_width * intake_coridor_height + 3.1, 2)
section_6 = round(2.5 * 5., 2)

exhaust_coridor_width = 3.3
exhaust_coridor_height = 2.45

section_8 = round(exhaust_coridor_width * exhaust_coridor_height + 3.8 * 4.4, 2)
section_9 = round(exhaust_coridor_width * exhaust_coridor_height + 6.2 * 4.5, 2)
section_10 = round(exhaust_coridor_width * exhaust_coridor_height + 6.2 * 4.7, 2)
section_11 = round(exhaust_coridor_width * exhaust_coridor_height + 6.2 * 4.9, 2)
section_12 = round(exhaust_coridor_width * exhaust_coridor_height + 9.1 * 5.0, 2)
section_13 = round(exhaust_coridor_width * exhaust_coridor_height + 11.8 * 5.1, 2)
section_14 = round(exhaust_coridor_width * exhaust_coridor_height + 15.2 * 5.1, 2)


list_sections = {}
list_sections["section_0"] = [section_0, G_sec_0, v_air(G_sec_0, section_0)]
list_sections["section_1"] = [section_1, G_sec_1, v_air(G_sec_1, section_1)]
list_sections["section_2"] = [section_2, G_sec_2, v_air(G_sec_2, section_2)]
list_sections["section_3"] = [section_3, G_sec_3, v_air(G_sec_3, section_3)]
list_sections["section_4"] = [section_4, G_sec_4, v_air(G_sec_4, section_4)]
list_sections["section_5"] = [section_5, G_sec_5/2, v_air(G_sec_5/2, section_5)]
list_sections["section_6"] = [section_6, G_sec_5/2, v_air(G_sec_5/2, section_6)]
# list_sections["section_6"] = [section_6, G_sec_6, v_air(G_sec_6, section_6)]
# list_sections["section_7"] = [section_7, G_sec_7, v_air(G_sec_7, section_7)]
# list_sections["section_71"] = [section_71, G_sec_71, v_air(G_sec_71, section_71)]
list_sections["section_8"] = [section_8, G_sec_8, v_air(G_sec_8, section_8)]
list_sections["section_9"] = [section_9, G_sec_9, v_air(G_sec_9, section_9)]
list_sections["section_10"] = [section_10, G_sec_10, v_air(G_sec_10, section_10)]
list_sections["section_11"] = [section_11, G_sec_11, v_air(G_sec_11, section_11)]
list_sections["section_12"] = [section_12, G_sec_12, v_air(G_sec_12, section_12)]
list_sections["section_13"] = [section_13, G_sec_13, v_air(G_sec_13, section_13)]
list_sections["section_14"] = [section_14, G_sec_14, v_air(G_sec_14, section_14)]




for key in list_sections:
    print(f" {key} - {list_sections[key][0]} m2, расход - {list_sections[key][1]} м3/ч, скорость - {list_sections[key][2]} м/с")

```

```python
from matplotlib import pyplot as plt 
from matplotlib import figure 

x = [i for i in list_sections.keys()]
y = []
flow = []

for i in range(len(list_sections)):
  temp = x[i]
  y.append(list_sections[temp][2])
  flow.append(list_sections[temp][1])

print(x)
print(y)

plt.figure(figsize=(12, 6), dpi=80)

for i in range(len(x)):
  plt.text(x[i],y[i], f"{y[i]} м/с")
  plt.text(x[i],y[i]-0.1, f"{flow[i]} м3/ч")

plt.xticks(rotation=90)
plt.plot(x, y, "-")
plt.title("sections")
plt.xlabel('sections')
plt.ylabel('velocity, m/s')
plt.grid(True)
# plt.fill_between(x, y, 2, facecolor='g', alpha=0.5)

plt.show()
```
